```python
# Importation
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt; plt.rcdefaults()
import matplotlib.pyplot as plt
from pylab import bar, show, plot
import seaborn as sns
import timeit
```

```python
# We read the file
df=pd.read_csv('b-way.csv')
```

```python
# PLot #1; this one is about the Years and total accumulation of money!
sns.jointplot(x='Year',y='Gross',data=df)
```

```python
# This is just a plot that distributes the accumulation of money graph
sns.distplot(df["Gross"],kde=False)
```

```python
# WE're making a box plot using attendance and years
sns.boxplot(x='Year',y='Attendance',data=df)
```

```python
# A swarmplot of the New Productions over the years
sns.swarmplot(x='New_Productions',y='Year',data=df)
```

```python
# We are making a coorelation heatmap!
qua = df.corr()
sns.heatmap(qua,cmap='coolwarm')
```

```python
# We are making a facet grid of the columns 
g = sns.FacetGrid(data=df, col='Gross')
g.map(sns.distplot, 'Attendance',kde=False)
```
