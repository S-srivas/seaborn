```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('New Data.csv', parse_dates=['Year'], index_col='Year')
```

```python
def plot_df(df, x, y, title="", xlabel='Year', ylabel='Rape', dpi=100):
    plt.figure(figsize=(16,5), dpi=dpi)
    plt.plot(x, y, color='tab:purple')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.xticks(rotation=90)
    plt.show()

plot_df(df, x=df.index, y=df.Rape, title='Rape by year')  
```
